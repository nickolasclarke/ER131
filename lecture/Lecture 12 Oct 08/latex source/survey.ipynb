```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.ticker import MaxNLocator
```

```python
results = pd.read_csv('ER131 2019 Survey.csv')
results.head()
```

```python
results.loc[results['The lab notebooks are...']== 'about the right level of difficulty','The lab notebooks are...'] = 'about right'
results.loc[results['The lab notebooks are...']== 'about the right level of difficulty','The lab notebooks are...'] = 'about right'
results.loc[results['The reading is...']== 'too much to handle','The reading is...'] = 'too much'
```

```python
sns.set(rc={'figure.figsize':(2,5)})
pace = sns.countplot(x = 'The lecture pace is...', data = results, orient = "h")
pace.yaxis.set_major_locator(MaxNLocator(integer=True))

pace.set_title('Lecture pace...',fontsize=18)
pace.set_xticklabels(pace.get_xticklabels(), fontsize=14)
pace.set_ylabel('responses', fontsize = 14)


pace.set_xticklabels(pace.get_xticklabels(), rotation=90)
pace.set_xlabel('')
pacefig = pace.get_figure()
pacefig.savefig("survey_pace.png", dpi=300, bbox_inches='tight')
```

```python
labbook = sns.countplot(x = 'The lab notebooks are...', data = results)
labbook.yaxis.set_major_locator(MaxNLocator(integer=True))
labbook.set_title('The lab \n notebooks are...',fontsize=18)


labbook.set_title('Lab notebooks...',fontsize=18)
labbook.set_xticklabels(labbook.get_xticklabels(), fontsize=14)
labbook.set_ylabel('responses', fontsize = 14)

labbook.set_xlabel('')
labbook.set_xticklabels(labbook.get_xticklabels(), rotation=90)
labbookfig = labbook.get_figure()
labbookfig.savefig("survey_lab_notebooks.png", dpi=300, bbox_inches = 'tight')
```
