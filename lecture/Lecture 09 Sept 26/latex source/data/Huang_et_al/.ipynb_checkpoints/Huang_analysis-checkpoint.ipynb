```python
from numpy import random
```

```python
np.random.seed(15)
x = np.random.randn(10)*7+ 20
#x = np.arange(5,30,2.5)
lolim = 300
hilim = 7500
vel = 300
ht = 500
y = -0.5*9.8*x**2 + vel*x + ht
```

```python
plt.scatter(x,y, s = sz)

SMALL_SIZE = 14
MEDIUM_SIZE = 14
BIGGER_SIZE = 16

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)
plt.ylim(lolim, hilim)

plt.savefig('projectile_notitle.pdf', format = 'pdf', dpi = 300, bbox_inches='tight')
```

```python
plt.scatter(x,y, s = sz)
plt.title('Projectile motion')
plt.ylabel('Height (m)')
plt.xlabel('Time (s)')

x_more = np.arange(np.min(x), np.max(x), 0.01)
y_more = -0.5*9.8*x_more**2 + vel*x_more + ht

plt.plot(x_more,y_more,'-', c = 'r')

SMALL_SIZE = 14
MEDIUM_SIZE = 14
BIGGER_SIZE = 16

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)
plt.ylim(lolim, hilim)

plt.savefig('projectile.pdf', format = 'pdf', dpi = 300, bbox_inches='tight')
```

```python
eps = np.random.randn(10)*500
y = -0.5*9.8*x**2 + vel*x + ht + eps
plt.scatter(x,y, s = sz)
plt.title('Noisy projectile motion')
plt.ylabel('Height (m)')
plt.xlabel('Time (s)')

plt.plot(x_more,y_more,'-', c = 'r')

SMALL_SIZE = 14
MEDIUM_SIZE = 14
BIGGER_SIZE = 16

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)
plt.ylim(lolim, hilim)

plt.savefig('projectile_noise.pdf', format = 'pdf', dpi = 300, bbox_inches='tight')
```

```python
from sklearn import neighbors
```

```python
pm25_array = np.array(seoul['pm2.5']).reshape(-1,1) # knn algorithm wants a NP array of this shape.
```

```python
knn_reg = neighbors.KNeighborsRegressor(n_neighbors = 100, weights='uniform')
knn_reg.fit(pm25_array, seoul['circ'])

X_pred = np.arange(0,120,0.01).reshape(-1,1)
output = knn_reg.predict(X_pred) #use the model we just created to predict

sns.regplot(seoul['pm2.5'], seoul['circ'], line_kws={'color': 'red'}, ci = None)

plt.plot(X_pred, output,'-', c = 'y')

plt.xlabel('PM2.5 concentrations (micrograms / cubic meter)')
plt.ylabel('Cardiovascular disease visits')
plt.title('Daily ER visits, Korean Hospitals, 2010-2013')

SMALL_SIZE = 14
MEDIUM_SIZE = 14
BIGGER_SIZE = 16

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)
plt.savefig('huang_pm25vcirc_knn.pdf', format = 'pdf', dpi = 300)
```
