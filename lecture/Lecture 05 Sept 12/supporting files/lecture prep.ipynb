```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dat = pd.read_csv('Electricity generation by fuel.csv', header = 6)
```

```python
dat
```

```python
dat.drop(6, axis = 0, inplace = True)
dat
```

```python
dat.index = dat['Year']
dat.drop('Year', axis = 1, inplace = True)
```

```python
dat
```

```python
dat.drop(['Unnamed: 13'], axis = 1, inplace = True)
```

```python
dat_tr = dat.transpose()
```

```python
dat_tr
```

```python
dat_tr['Source'] = dat_tr.index
dat_tr
```

```python
yr = '2015'
g = sns.barplot(x='Source', y = yr, data = dat_tr)
g.set_xticklabels(g.get_xticklabels(), rotation=45)
plt.ylabel('GWh')
plt.title('Global Electricity production by fuel in '+ yr)
plt.tight_layout()
plt.savefig(yr+'ElectricityProduction.png', dpi = 300)
```
