This notebook works with data from the Huang et al paper in the same directory.  Hospital visits for respiratory problems, for cardiac problems, and pollution data.  

```python
import pandas as pd
import numpy as np
```

```python
seoul = pd.read_csv('data_20170517.csv')
```

```python
seoul.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(seoul['circ_F'], seoul['circ_M'], c = seoul['meantemp'], cmap = 'viridis')
cbar = plt.colorbar();  # show color scale
cbar.set_label('temperature (C)')
plt.xlabel('PM2.5 concentrations (micrograms / cubic meter)')
plt.ylabel('Daily ER visits for cardiovascular disease')
plt.title('')
```
